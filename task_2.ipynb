```python
# 1. Importing Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Load the Dataset (Kaggle Titanic)
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
# 3. Overview of the Dataset
print("First 5 rows of dataset:")
display(df.head())

print("\nDataset Info:")
df.info()

print("\nMissing Values:")
display(df.isnull().sum())

print("\nStatistical Summary:")
display(df.describe())
```

```python
# 4. Handling Missing Values
# Fill Age with median, Embarked with mode, drop Cabin due to too many missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns='Cabin', inplace=True)

print("\nMissing Values after cleaning:")
display(df.isnull().sum())
```

```python
# 5. Univariate Analysis (Single Variable)

# Gender Distribution
sns.countplot(data=df, x='Sex', palette='Set2')
plt.title("Gender Distribution")
plt.show()

# Survival Count
sns.countplot(data=df, x='Survived', palette='coolwarm')
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()

# Passenger Class
sns.countplot(data=df, x='Pclass', palette='pastel')
plt.title("Passenger Class Distribution")
plt.show()

# Age Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# 6. Bivariate Analysis (Comparing Variables)

# Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set1')
plt.title("Survival by Gender")
plt.show()

# Survival by Passenger Class
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set3')
plt.title("Survival by Passenger Class")
plt.show()

# Boxplot: Age vs Survived
sns.boxplot(data=df, x='Survived', y='Age', palette='Set2')
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# 7. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 8. Key Insight Example
survival_rate = df['Survived'].mean() * 100
print(f"\nOverall Survival Rate: {survival_rate:.2f}%")

survival_by_gender = df.groupby('Sex')['Survived'].mean() * 100
print("\nSurvival Rate by Gender:")
display(survival_by_gender)

survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print("\nSurvival Rate by Passenger Class:")
display(survival_by_class)
```
